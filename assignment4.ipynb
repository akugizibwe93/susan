```python
#1. Data Preparation
#importing 
import pandas as pd
import numpy as np
import os
# Specify the current and new file names
current_file_name = "E:\\Nexford-Python-Assignments2026\\module 4-assignment\\netflix_data.csv"
new_file_name = "E:\\Nexford-Python-Assignments2026\\module 4-assignment\\Netflix_shows_movies.csv"
# Rename the file
os.rename(current_file_name, new_file_name) 
```

```python
#view the contents of the file
df = pd.read_csv ("E:\\Nexford-Python-Assignments2026\\module 4-assignment\\Netflix_shows_movies.csv")
df
```

```python
#Data cleaning
# check total number of missing values per column:
df.isnull().sum()
```

```python
# Check for missing values using isna()
print(df.isna())
```

```python
#replace all missing fields in the column 'director' with 'no director' value
df['director'] = df['director'].fillna('no director')
```

```python
#replace all missing fields in the column 'cast' with 'not listed' value
df['cast'] = df['cast'].fillna('not listed')
```

```python
#fill all values in country column with 'not known' value
df['country'] = df['country'].fillna('not known')
```

```python
#fill all values in date_added column with 'not listed' value
df['date_added'] = df['date_added'].fillna('not listed')
```

```python
#drop all rows in the column rating that have no values
df = df.dropna(subset=['rating'])
```

```python
# verify that all missing values have been handled in all columns:
df.isnull().sum()
```

```python
#save the changes to the csv file
df.to_csv("E:\\Nexford-Python-Assignments2026\\module 4-assignment\\Netflix_shows_movies.csv", index=False)
```

```python


df = pd.read_csv ("E:\\Nexford-Python-Assignments2026\\module 4-assignment\\Netflix_shows_movies.csv")
# Split genres and count
genre_series = df['listed_in'].str.split(', ').explode()
genre_counts = genre_series.value_counts().head(15)

plt.figure(figsize=(10,6))
sns.barplot(x=genre_counts.values, y=genre_counts.index, color='deeppink')
plt.title('Most Watched Genres (Top 15)')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

```python

# --- RATINGS DISTRIBUTION ---
rating_counts = df['rating'].value_counts()

plt.figure(figsize=(10,6))
sns.barplot(x=rating_counts.index, y=rating_counts.values, color='green')
plt.title('Ratings Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
