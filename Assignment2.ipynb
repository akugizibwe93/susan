```python
4 # Error Handling:
import csv
from typing import Dict

def load_employees_dict(csv_path: str = "Total.csv",
                        name_col: str = "EmployeeName") -> Dict[str, dict]:
    """
    Read salary data from CSV into a dictionary keyed by EmployeeName.
    Gracefully handles:
      - Missing file / permission issues
      - Missing required column
      - Empty or malformed CSV
      - Encoding problems
      - Empty rows
    """
    employees: Dict[str, dict] = {}

    try:
        # Open with a tolerant decoder (replace undecodable chars rather than fail)
        with open(csv_path, "r", encoding="utf-8", errors="replace", newline="") as file:
            reader = csv.DictReader(file)

            # Validate header
            if not reader.fieldnames:
                raise ValueError("No header row found; the CSV may be empty or corrupted.")

            if name_col not in reader.fieldnames:
                raise KeyError(
                    f"Required column '{name_col}' not found. "
                    f"Available columns: {reader.fieldnames}"
                )

            # Build the dictionary (first occurrence wins; change policy if needed)
            for row in reader:
                # Skip completely empty lines
                if not any(row.values()):
                    continue

                # Normalize string values (strip whitespace)
                clean_row = {k: (v.strip() if isinstance(v, str) else v) for k, v in row.items()}

                name = clean_row.get(name_col, "")
                if not name:
                    # Row has no usable EmployeeName — skip but continue processing
                    print("Skipping row with missing EmployeeName:", clean_row)
                    continue

                # Keep the first record for a duplicate name; use `employees[name] = clean_row`
                # if you prefer "last occurrence wins"
                if name not in employees:
                    employees[name] = clean_row

        print(f" Loaded {len(employees)} employees from '{csv_path}'.")
        return employees

    except FileNotFoundError:
        print(f" File not found: {csv_path}. Check the path or place the file next to your notebook.")
    except PermissionError:
        print(" Permission denied. Close the file if it's open in another application (e.g., Excel) and retry.")
    except UnicodeDecodeError:
        print(" Encoding error while reading the CSV. Try opening with encoding='latin-1'.")
    except csv.Error as e:
        print(f" CSV parsing error: {e}")
    except KeyError as e:
        print(f" {e}")
    except ValueError as e:
        print(f" {e}")
    except Exception as e:
        print(f" Unexpected error: {e}")

    # Safe default on any failure
    return {}


employees = load_employees_dict("Total.csv", name_col="EmployeeName")

# Preview the first 10 entries without flooding the output (the file is very large)
for i, (k, v) in enumerate(employees.items()):
    if i == 10:
        break
    print(k, ":", v)
```

```python
5. # Export Employee Details
import pandas as pd
from pathlib import Path
import zipfile, io, re

# If you haven't already, import the CSV (no error handling as requested)
# df = pd.read_csv("Total.csv")  # columns verified in your upload: EmployeeName, JobTitle, BasePay, OvertimePay, OtherPay, Benefits, TotalPay, TotalPayBenefits, Year

def safe_filename(text: str) -> str:
    """Create a filesystem-friendly filename from any text."""
    return re.sub(r'[^A-Za-z0-9._-]+', '_', str(text)).strip('_') or 'employee'

def export_employee_to_zip(emp_name: str,
                           df: pd.DataFrame,
                           output_zip: Path = Path("Employee Profile.zip"),
                           folder_name: str = "Employee Profile",
                           name_col: str = "EmployeeName",
                           filename_prefix: str = "employee") -> Path:
    """
    Find `emp_name` in df[name_col], save their row to CSV,
    and write it inside `output_zip` at: <folder_name>/<filename_prefix>_<name>.csv
    - Exact, case-insensitive match
    - If multiple rows match, the first match is exported (you can change this policy)
    """

    # 1) Locate the employee (exact, case-insensitive)
    series = df[name_col].astype(str)
    mask = series.str.casefold() == str(emp_name).strip().casefold()
    matches = df[mask]

    if matches.empty:
        print(f"Employee not found: {emp_name}")
        return output_zip

    # 2) Take the first match and convert to CSV in-memory
    row_dict = matches.iloc[0].to_dict()
    csv_buf = io.StringIO()
    pd.DataFrame([row_dict]).to_csv(csv_buf, index=False, encoding="utf-8")
    csv_bytes = csv_buf.getvalue().encode("utf-8")

    # 3) Build internal path: Employee Profile/employee_<EMPLOYEE_NAME>.csv
    file_name = f"{filename_prefix}_{safe_filename(row_dict.get(name_col, emp_name))}.csv"
    internal_path = f"{folder_name}/{file_name}"

    # 4) Append to (or create) the ZIP
    with zipfile.ZipFile(output_zip, mode="a", compression=zipfile.ZIP_DEFLATED) as zf:
        zf.writestr(internal_path, csv_bytes)

    print(f" Exported: {output_zip} → {internal_path}")
    return output_zip
    export_employee_to_zip("NATHANIEL FORD", df)
```
